Python's `datetime` module provides various formats to represent and manipulate date and time. You can use the `strftime` (string format time) and `strptime` (string parse time) methods for formatting and parsing date/time objects.

---

### **Common Date and Time Formats**

| **Format Code** | **Meaning**                          | **Example**           |
|------------------|--------------------------------------|-----------------------|
| **`%Y`**         | Year with century                   | `2024`               |
| **`%y`**         | Year without century (2 digits)     | `24`                 |
| **`%m`**         | Month as zero-padded decimal        | `01` to `12`         |
| **`%B`**         | Full month name                     | `December`           |
| **`%b`**         | Abbreviated month name              | `Dec`                |
| **`%d`**         | Day of the month (zero-padded)      | `01` to `31`         |
| **`%A`**         | Full weekday name                   | `Monday`             |
| **`%a`**         | Abbreviated weekday name            | `Mon`                |
| **`%H`**         | Hour (24-hour clock, zero-padded)   | `00` to `23`         |
| **`%I`**         | Hour (12-hour clock, zero-padded)   | `01` to `12`         |
| **`%p`**         | AM or PM                            | `AM` or `PM`         |
| **`%M`**         | Minute (zero-padded)                | `00` to `59`         |
| **`%S`**         | Second (zero-padded)                | `00` to `59`         |
| **`%f`**         | Microsecond                         | `000000` to `999999` |
| **`%z`**         | UTC offset                          | `+0000`, `-0530`     |
| **`%Z`**         | Timezone name                       | `UTC`, `IST`         |
| **`%j`**         | Day of the year (1-366)             | `001` to `366`       |
| **`%U`**         | Week number of the year (Sunday as the first day) | `00` to `53`  |
| **`%W`**         | Week number of the year (Monday as the first day) | `00` to `53`  |

---

### **Examples**

#### Format Current Date and Time:
```python
from datetime import datetime

# Get current date and time
now = datetime.now()

# Format the date and time
formatted = now.strftime("%Y-%m-%d %H:%M:%S")
print("Formatted Date and Time:", formatted)
```
**Output:**
```
Formatted Date and Time: 2024-12-16 14:30:45
```

---

#### Parse Date from String:
```python
from datetime import datetime

# Parse date string into datetime object
date_str = "16-12-2024"
parsed_date = datetime.strptime(date_str, "%d-%m-%Y")
print("Parsed Date:", parsed_date)
```
**Output:**
```
Parsed Date: 2024-12-16 00:00:00
```

---

#### Custom Formats:
```python
from datetime import datetime

now = datetime.now()

# Day of the week
print("Day of the Week:", now.strftime("%A"))

# Month and Year
print("Month and Year:", now.strftime("%B %Y"))

# ISO 8601 Format
print("ISO Format:", now.strftime("%Y-%m-%dT%H:%M:%S"))
```
**Output:**
```
Day of the Week: Monday
Month and Year: December 2024
ISO Format: 2024-12-16T14:30:45
```
